# 1. personal mbti general community crawler

```python
'''412페이지까지 존재'''
base_url = 'https://www.personalitycafe.com/general-chat/index{}.html'
```

```python
title = []
community_url = []
for i in range(1,412):
    if i ==1:
        print('1 번째 페이지 수집하고 있습니다.')
        url = base_url.format(i)
        res = requests.get(url)

        soup = BeautifulSoup(res.content,'html.parser')

        thread = soup.find('form',{'id':'thread_inlinemod_form'})

        thr2 = thread.find('ol',{'id':'stickies'})
        thr = thread.find('ol',{'id':'threads'})
        
        thr_ls0 = thr2.find_all('li',{'class':re.compile('threadbit')})[1:-1]
        thr_ls = thr.find_all('li',{'class':re.compile('threadbit')})
        
        for j in range(len(thr_ls0)):
            title0 = thr_ls0[j].find('h3',{'class':'threadtitle'}).get_text(strip=True)
            community_url0 = thr_ls0[j].find('h3',{'class':'threadtitle'}).find('a')['href']
            
            title.append(title0)
            community_url.append(community_url0)
            
        for k in range(len(thr_ls)):
            title0 = thr_ls[k].find('h3',{'class':'threadtitle'}).get_text(strip=True)
            community_url0 = thr_ls[k].find('h3',{'class':'threadtitle'}).find('a')['href']
            
            title.append(title0)
            community_url.append(community_url0)
            
    else:
        print(str(i)+' 페이지 수집하고 있습니다.')
        url = base_url.format(i)
        res = requests.get(url)

        soup = BeautifulSoup(res.content,'html.parser')

        thread = soup.find('form',{'id':'thread_inlinemod_form'})

        thr = thread.find('ol',{'id':'threads'})
        
        thr_ls = thr.find_all('li',{'class':re.compile('threadbit')})
        
        for k in range(len(thr_ls)):
            title0 = thr_ls[k].find('h3',{'class':'threadtitle'}).get_text(strip=True)
            community_url0 = thr_ls[k].find('h3',{'class':'threadtitle'}).find('a')['href']

            title.append(title0)
            community_url.append(community_url0)

            
print('수집완료했습니다.')

mbti_general_ls0 = [title,community_url]

mbti_general_ls = list(map(list,zip(*mbti_general_ls0)))

mbti_general_df = pd.DataFrame(mbti_general_ls,columns = ['title','url'])
```

```python
mbti_general_df.to_pickle('mbti_general.pkl')
```

# personal cafe 게시글 url 모두 수집

```python
'''support-suggestion, advice-center은 셀레늄으로 수집 해야함'''
plus_url_ls = ['general-psychology/index{}.html','trends-forum/index{}.html',
               'art-museum/index{}.html','book-music-movie-reviews/index{}.html',
              'education-career-talk/index{}.html','science-technology/index{}.html',
              'health-fitness/index{}.html','advice-center/index{}.html',
              'mafia/index{}.html','support-suggestions/index{}.html',
              'myers-briggs-forum/index{}.html','cognitive-functions/index{}.html',
              'socionics-forum/index{}.html','enneagram-personality-theory-forum/index{}.html']
```

```python
base_url = 'https://www.personalitycafe.com/'
community_url = []
title=[]
forum = []
for plus in plus_url_ls[-4:]:
    print(plus.split('/')[0]+'forum 수집하고 있습니다.')
    fin_url = base_url + plus.format(1)
    
    res = requests.get(fin_url)
    soup = BeautifulSoup(res.content,'html.parser')
    page_url0 = soup.find('span',{'class':'first_last'}).find('a')['href']
    page_num = ''.join(re.findall('[0-9]',page_url0))
    
    for page in list(range(1,int(page_num)+1)):
        fin_url = base_url + plus.format(page)
        if page ==1:
            print('1 번째 페이지 수집하고 있습니다.')
            res = requests.get(fin_url)

            soup = BeautifulSoup(res.content,'html.parser')

            thread = soup.find('form',{'id':'thread_inlinemod_form'})

            thr2 = thread.find('ol',{'id':'stickies'})
            thr = thread.find('ol',{'id':'threads'})

            thr_ls0 = thr2.find_all('li',{'class':re.compile('threadbit')})[1:-1]
            thr_ls = thr.find_all('li',{'class':re.compile('threadbit')})

            for j in range(len(thr_ls0)):
                title0 = thr_ls0[j].find('h3',{'class':'threadtitle'}).get_text(strip=True)
                community_url0 = thr_ls0[j].find('h3',{'class':'threadtitle'}).find('a')['href']

                title.append(title0)
                community_url.append(community_url0)
                forum.append(plus.split('/')[0])

            for k in range(len(thr_ls)):
                title0 = thr_ls[k].find('h3',{'class':'threadtitle'}).get_text(strip=True)
                community_url0 = thr_ls[k].find('h3',{'class':'threadtitle'}).find('a')['href']

                title.append(title0)
                community_url.append(community_url0)
                forum.append(plus.split('/')[0])

        else:
            print(str(page)+' 페이지 수집하고 있습니다.')
            res = requests.get(fin_url)

            soup = BeautifulSoup(res.content,'html.parser')

            thread = soup.find('form',{'id':'thread_inlinemod_form'})

            thr = thread.find('ol',{'id':'threads'})

            thr_ls = thr.find_all('li',{'class':re.compile('threadbit')})

            for k in range(len(thr_ls)):
                title0 = thr_ls[k].find('h3',{'class':'threadtitle'}).get_text(strip=True)
                community_url0 = thr_ls[k].find('h3',{'class':'threadtitle'}).find('a')['href']

                title.append(title0)
                community_url.append(community_url0)
                forum.append(plus.split('/')[0])
    print(plus.split('/')[0]+'forum 수집 완료했습니다.')
                

print('전체 forum 수집완료했습니다.')
```

```python
mbti_general_ls0 = [title,community_url,forum]

mbti_general_ls = list(map(list,zip(*mbti_general_ls0)))

mbti_general_df = pd.DataFrame(mbti_general_ls,columns = ['title','url','forum'])
```

```python
mbti_general_df.to_pickle('mbti_plus_url3.pkl')
```
